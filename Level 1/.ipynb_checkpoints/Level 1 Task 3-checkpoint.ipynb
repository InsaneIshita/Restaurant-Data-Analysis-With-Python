### Task 3 : Price Range Distribution

* Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
* Calculate the percentage of restaurants in each price range category.

### Loading necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the dataset

```python
url = "https://drive.google.com/uc?export=download&id=1dKZxapT3xLzOTOpy1LCyqctL8YEvzp4Y"
df = pd.read_csv(url)
#previewing first few rows of the dataset
df.head()
```

```python
# Checking the total number of rows and columns
df.shape
```

### Visualizing the distribution of Price ranges among the restaurants

```python
#creating a histogram to see the distribution of price range vs. number of restaurants
plt.figure(figsize=(8, 6))
sns.histplot(data = df , x = df["Price range"] , bins = 4 , hue = df["Price range"] , palette = "dark" , edgecolor = "white")
plt.xlabel("Price Range" , fontsize = 15)
plt.ylabel("Number of restaurants" , fontsize = 15)
plt.title("Distribution of Price Ranges among the restaurants" , fontsize = 19)

# adding data labels
price_counts = df['Price range'].value_counts().sort_index()
for index, (price, value) in enumerate(price_counts.items()):
    plt.text(price, value + 50, str(value), ha='left')
   
plt.show()
```

### Calculating the percentage of restaurants in each price range category

```python
#calculating the percentages of restaurants per price range 
price_percentages = round((price_counts/price_counts.sum())*100,2)

#visualizing using pie chart
plt.figure(figsize=(8, 6))
c = ["brown","darkolivegreen","goldenrod","royalblue"]
ex = [0,0,0,0.1]
plt.pie(price_percentages.values , labels = price_percentages.index , autopct = "%.2f%%" ,  colors = c , explode = ex)
plt.title("Percentages of Restaurants in each Price Range" , fontsize = 19)

plt.show()
```
